```python
#panda nos permite manipular el archivo csv https://pandas.pydata.org
import pandas as pd
#numpy nos permite hacer operaciones usando vectores https://numpy.org
import numpy as np
#plotly nos permite crear graficas interactivas https://plotly.com/python/
import plotly.express as px
%matplotlib inline
#usamos la liga que está directamente en la página. 
particles = pd.read_csv('https://opendata.cern.ch/record/5208/files/Zmumu.csv',delimiter=',')
particles.head()
```

```python
# fig_5=px.histogram(particles['pt1'],particles['pt1'])
# fig_5.show()
```

Se busca obtener una visualización para el espectro de masa invariante dimuónica.
- Los datos contienen eventos donde se registraron 2 muones.
- Para aseguranos de que sean una pareja, éstos deben tener carga opuesta. Podemos introducir un filtro más;


```python
df_2m = particles.query("Q1 != Q2")
#print(df_2m)
```

- También es necesario computar la masa invariante. La expresión a utilizar es:
$$M^2 = 2p_{T1}p_{T2}(\cosh(\eta_1-\eta_2)-\cos(\phi_1-\phi_2))$$

- $\eta_{1,2},\quad pT_{1,2},\quad \phi_{1,2} \quad$ son todas columnas en la hoja de datos. Ya tenemos numpy, entonces podemos formar la expresión:


```python
masa_invariante = np.sqrt(2*df_2m.pt1*df_2m.pt2*(np.cosh(df_2m.eta1-df_2m.eta1)-np.cos(df_2m.phi1-df_2m.phi2)))
#print(masa_invariante)
```

Podemos crear el histograma. Como ya tenemos plotly express,

```python
fig1 = px.histogram(data_frame=df_2m, x=masa_invariante,log_y= True, width=700, height=700,)
fig1.update_xaxes(title_text='Masa invariante (GeV/c^2)')
fig1.update_yaxes(title_text='Eventos/MeV')
fig1.show()
numero_eventos=len(df_2m)
print(numero_eventos)
```

Histograma para muones (sin distinción de carga)

```python
df = particles.head()
fig2 = px.histogram(df, x=masa_invariante,log_y= True, width=700, height=700,)
fig2.update_xaxes(title_text='Masa invariante (GeV/c^2)')
fig2.update_yaxes(title_text='Eventos/MeV')
fig2.show()
numero_eventos1= len(particles)
print(numero_eventos1)
```
